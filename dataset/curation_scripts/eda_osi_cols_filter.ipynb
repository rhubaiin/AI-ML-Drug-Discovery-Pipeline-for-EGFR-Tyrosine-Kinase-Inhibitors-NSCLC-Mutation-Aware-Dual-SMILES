```python
import pandas as pd

df = pd.read_csv('/mnt/d/Publications/project_insilico/dataset/osimertinib/osi_bioactivity_chembl.csv')
df['Assay Description'] = df['Assay Description'].str.lower().str.strip()

pattern = (
    # --- EGFR and General Terms ---
    r'\b(egfr|epidermal\s+growth\s+factor\s+receptor)\b|'  
    r'\b(wild|mutant|mutations?)\b|'  
    
    # --- Exon-Specific Mutations ---
    r'exon\s?(?:18|19|20|21)\s?(?:ins|del|mutations?|insertions?|deletions?)\b|'
    r'\b(del|ins)(?:18|19|20|21)\b|'  
    
    # --- General Mutation Types ---
    r'\b(deletions?|insertions?|substitutions?|del|ins|sub)\b|'  
    
    # --- Specific Common Mutations ---
    r'\b([A-Z]\d{3,4}[A-Z]|T790M|L858R|C797S|L861Q|G719X)\b|'
    
    # --- Cell Lines and Other Terms ---
    r'\b(baf3|nsclc)\b|'
    r'\b(A-431|HCC4006|U-87\s?MG|PC-9|CL97|Calu-3|HCC827)\b|'
    r'\bNCI-(?:H1666|1155|H1975|H3255|H1299|H2405|H1650|H1651)\b'
)

filtered_df = df[df['Assay Description'].str.contains(pattern, na=False, regex=True, case=False)]
print(filtered_df.head(20))
```

```python
import pandas as pd
import numpy as np

# Load your DataFrame
df = pd.read_csv(osi_df_cols)


# Clean the Target Name column
df['Target Name'] = df['Target Name'].str.strip().replace('', np.nan)
df['Assay Description'] = df['Assay Description'].str.lower().str.strip()

# Part A: EGFR-related targets and cell lines to keep
targets_to_keep = [
    'Epidermal growth factor receptor erbB1',
    'EGFR', 'EGFR tyrosine kinase', 'EGFR TKD', 'EGFR kinase domain',
    'EGFR L858R', 'EGFR T790M', 'EGFR exon 19 deletion',
    # EGFR-mutant cell lines
    'A-431', 'HCC4006', 'U-87 MG', 'NCI-H1666', 'NCI-H1975', 'HCC827',
    'NCI-H3255', 'PC-9', 'NCI-H1650', 'CL97', 'Calu-3',
    # Additional EGFR-relevant cell lines
    'NCI-H820', 'NCI-H3255', 'HCC827GR5', 'PC-9/BRC1'
]

# Part B: Cytotoxicity / ADME-Tox to filter out
cyto_cells = [
    'hepatocyte', 'cho', 'hk-2', 'l02', 'hela', 'huvec', 'bj', 'ea.hy 926',
    'beas-2b', 'hepg2', 'ht-29', 'kg-1', 'lymphoma cell line', 'calu-6',
    'nci-h292', 'mda-mb-468', 'mda-mb-231', 'vero', 'vero c1008',
    'plasma', 'liver', 'primary hepatocyte', 'hep rg', 'hepatocytes',
    # species
    'mus musculus', 'mouse', 'rattus norvegicus', 'rat', 'cavia porcellus', 'guinea pig',
    # general cancer panels
    'a549', 'mcf7', 'sk-br-3', 't47d', 'bt-474', 'hct-116', 'lovo', 
    'dld-1', 'sw480', 'sw-620', 'panc-1', 'ovcar-8', 'sk-ov-3', 'kb', 
    'a101d', 'a498', 'nb69', 'rmg-i', 'l5178y', 'nci-h322m', 'nci-h510a', 
    'nci-h838', 'nci-h2170'
]

cyto_markers = [
    'herg', 'potassium channel', 'cardiac ion channel',
    'cytochrome p450', 'cyp', 'cyp1a2', 'cyp2c9', 'cyp2d6', 'cyp2c8', 
    'cyp3a4', 'cyp2b6', 'cyp2c19', 'cyp3a5',
    'monoamine oxidase', 'mao', 'cyclooxygenase', 'cox-1', 'cox-2',
    'admet', 'no relevant target', 'non-protein target', 
    'molecular identity unknown', 'unspecific', 'off-target',
    # safety receptors
    'beta-1 adrenergic receptor', 'beta-2 adrenergic receptor',
    'type-1 angiotensin ii receptor', 'fibroblast growth factor receptor',
    'adenosine a1 receptor', 'adenosine a2a receptor', 'adenosine a3 receptor',
    '5-ht', 'serotonin receptor', 'dopamine receptor', 'histamine receptor',
    'opioid receptor', 'sodium channel', 'calcium channel'
]

# Part C: Non-EGFR tyrosine kinases and unrelated targets to filter out
other_activity_targets = [
    # Kinases / signaling
    'PI3-kinase', 'PI3K', 'AKT', 'mTOR', 'phosphoinositide 3-kinase',
    'Tyrosine-protein kinase ABL', 'Bcr-Abl', 'BCR-ABL',
    'Stem cell growth factor receptor', 'c-Kit', 'KIT',
    'Fibroblast growth factor receptor', 'FGFR',
    'Hepatocyte growth factor receptor', 'MET', 'c-Met',
    'Tyrosine-protein kinase receptor RET', 'RET',
    'Tyrosine-protein kinase receptor FLT3', 'FLT3',
    'Tyrosine-protein kinase JAK', 'JAK1', 'JAK2', 'JAK3',
    'EML4-ALK', 'ALK', 'Anaplastic lymphoma kinase',
    'Serine/threonine-protein kinase B-raf', 'BRAF', 'B-raf',
    'GAK', 'Chk1', 'Chk2', 'PLK', 'Polo-like kinase',
    'Insulin receptor', 'Insulin-like growth factor I receptor', 'IGF-1R',
    'VEGFR', 'PDGFR', 'SRC', 'SYK', 'BTK', 'TEK',
    
    # Other receptors
    'Androgen receptor', 'Estrogen receptor', 'Progesterone receptor',
    
    # Misc / non-human
    'Quinolone resistance protein norA', 'Venezuelan equine encephalitis virus',
    
    # Ambiguous labels
    'Unchecked', 'DEL', 'Not specified', 'Unknown'
]

# Convert all to lowercase for case-insensitive matching
targets_to_keep_lower = [t.lower() for t in targets_to_keep]
cyto_cells_lower = [c.lower() for c in cyto_cells]
cyto_markers_lower = [m.lower() for m in cyto_markers]
other_activity_lower = [o.lower() for o in other_activity_targets]

# Function to check if a target should be kept
def should_keep_target(target):
    if pd.isna(target):
        return False
    
    target_lower = target.lower()
    
    # First check if it's in our keep list
    for keep_target in targets_to_keep_lower:
        if keep_target in target_lower:
            return True
    
    # Then check if it should be filtered out
    for cyto_cell in cyto_cells_lower:
        if cyto_cell in target_lower:
            return False
            
    for cyto_marker in cyto_markers_lower:
        if cyto_marker in target_lower:
            return False
            
    for other_target in other_activity_lower:
        if other_target in target_lower:
            return False
    
    # If none of the above, keep it (assuming it might be EGFR-related)
    return True

# Apply the filtering
df_filtered_target_name = df[df['Target Name'].apply(should_keep_target)]

df_filtered_assay_desc = df[df['Assay Description'].apply(should_keep_target)]

# Save the filtered DataFrame
df_filtered_target_name.to_csv('filtered_target_name.csv', index=False)
df_filtered_assay_desc.to_csv('filtered__assaydesc.csv', index=False)

print(f"Original dataset size: {len(df)}")
print(f"Filtered dataset size: {len(df_filtered_assay_desc)}")
print(f"Filtered dataset size: {len(df_filtered_target_name)}")
print(f"Filtered dataset size: {len(df_filtered_target_name)}")
```
